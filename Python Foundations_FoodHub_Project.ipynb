# Python Foundations Project: FoodHub Data Analysis


### Context

The number of restaurants in New York is increasing day by day. A food aggregator company FoodHub offers access to multiple restaurants through a single smartphone app.

The app allows the restaurants to receive a direct online order from a customer. The app assigns a delivery person from the company to pick up the order after it is confirmed by the restaurant. The delivery person then uses the map to reach the restaurant and waits for the food package. Once the food package is handed over to the delivery person, he/she confirms the pick-up in the app and travels to the customer's location to deliver the food. The delivery person confirms the drop-off in the app after delivering the food package to the customer. The customer can rate the order in the app. The food aggregator earns money by collecting a fixed margin of the delivery order from the restaurants.

### Objective

FoodHub company has stored the data of the different orders made by the registered customers in their online portal. This report analyses the data to get a fair idea about the demand of different restaurants which will help the company in enhancing their customer experience.Some of the key questions are answered below along with data analysis that will help the company to improve the business.

## Conclusion and Recommendations



### Delivery Time

The delivery time impacts the rating given by the customers.The higher the delivery time, the lower the rating.

Customer satisfaction can be increased by implementing measures to reduce the delivery time from the time the order is placed.
This can be shared with restaurants to reduce the food preparation time if possible. 
On the Foodhub side the pickup and delivery schedules can be reviewed to find solutions to reduce the delivery time.

Customer satisfaction means repeat business and increased revenue.It also leads to free marketing through word of mouth that results to new business.



### There are 147% more orders over the weekend compared to weekdays.

This surge over the weekends should be anticipated and measures put in place to ensure smooth operation of the service.
For instance,temporary delivery personnel may be hired over the weekend to ensure deliveries are done in the shortest time possible.



### American cuisine is the most popular followed by Japanese cuisine.American cuisine generated the most revenue.

To improve the business revenue,more American and Japanese cuisine restaurants may be added on the app to give the customers a greater variety to choose from.



### 39% of the orders were not rated.

To encourage customers to rate their order,promts or pop ups may be introduced on the app to request for rating before customers exit the app. This will encourage feedback from customers that would assist in improving the service.
